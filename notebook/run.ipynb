```python
wave = tts.tts("ภาษาไทย ง่าย มาก มาก",return_type="waveform")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(np.array(wave), sr=16000)
```

```python
IPython.display.Audio(np.array(wave), rate=16000) # load a NumPy array
```
